### 1. Perform Detailed exploratory analysis

#### a. Define & calculate high level metrics like (Total Revenue, Total quantity, Total products, Total categories, Total sellers, Total locations, Total channels, Total payment methods etc…) 

```python
# Total Revenue

print('Total Revenue =', round(order_items.price.sum(),0))

# Total Quantity

print('Total Quantity =', order_items.shape[0])

# Total Products

print('Total Products =', products.product_id.nunique())

# Total Categories

print('Total Categories =', products.product_category_name.nunique())

# Total Sellers

print('Total Sellers =', sellers.seller_id.nunique())

# Total Locations

print('Total Locations =', geo_loc.geolocation_zip_code_prefix.nunique())

# Total Payment Methods

print('Total Payment Methods =', order_payments.payment_type.nunique())
```

#### b. Understanding how many new customers acquired every month

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders.order_purchase_timestamp, format = '%m/%d/%Y %H:%M')
```

```python
orders['Year_month'] = orders.order_purchase_timestamp.apply(lambda x: pd.Timestamp.strftime(x , format = "%Y-%m"))
```

```python
#Merging customers and orders data

cust_orders = pd.merge(customers, orders, how = 'left', on = 'customer_id')
```

```python
cust_month = cust_orders.loc[:,['customer_unique_id','Year_month']]
```

```python
cust_month.drop_duplicates(subset = 'customer_unique_id', inplace = True)
```

```python
new_cust_monthly = cust_month.groupby('Year_month')['customer_unique_id'].count()
new_cust_monthly
```

```python
new_cust_monthly.plot( kind = 'bar', figsize = (18,5))
plt.ylabel('Count of Customers')
plt.title('New Customers Acquired Every Month')
plt.show()
```

#### c. Understand the retention of customers on month on month basis

```python
cust_monthly_count = cust_orders.groupby('Year_month')['customer_unique_id'].count().reset_index()
cust_monthly_count.head()
```

```python
retained_cust_monthly = pd.merge(new_cust_monthly,cust_monthly_count, how = 'left', on = 'Year_month')
retained_cust_monthly.head()
```

```python
retained_cust_monthly['retained_cust'] = retained_cust_monthly.customer_unique_id_y - retained_cust_monthly.customer_unique_id_x
retained_cust_monthly.head()
```

```python
retained_cust_monthly.plot(kind= 'bar', x= 'Year_month' , y = 'retained_cust' , figsize = (16,6))
plt.ylabel('Count of Customers')
plt.title('Retention of Customers on Monthly Basis')
plt.show()
```

#### d. How the revenues from existing/new customers on month on month basis

```python
order_item_month = pd.merge(order_items, orders, how = 'left',on = 'order_id')
order_item_month['net_price'] = order_item_month.price + order_item_month.freight_value
```

```python
revenue_monthly = order_item_month.groupby('Year_month')[['net_price']].sum()
revenue_monthly.head()
```

```python
revenue_monthly.plot( kind = 'bar', figsize =(16,6))
plt.ylabel('Total Amount')
plt.title('Total Revenue on Monthly Basis')
plt.show()
```

#### e. Understand the trends/seasonality of sales, quantity by category, location, month, week, day, time, channel, payment method etc…

```python
order_items_prod = pd.merge(order_item_month, products, how = 'left', on = 'product_id')
order_items_prod['month'] = order_items_prod.order_purchase_timestamp.apply(lambda x: pd.Timestamp.strftime(x , format = "%b"))
order_items_prod.head()
```

```python
# Seasonality of sales, quantity by category.

category_qty_sales = order_items_prod.groupby('product_category_name').agg({'product_id':'count','price':'sum'}).reset_index().rename(columns={'product_id':'quantity','price':'sales'})
category_qty_sales
```

```python
# Seasonality of sales, quantity by month.
month_qty_sales = order_items_prod.groupby('month').agg({'product_id':'count','price': 'sum'}).reset_index().rename(columns={'product_id':'quantity','price':'sales'})
month_qty_sales
```

```python
# Analyse sales, quantity by week. Add a new column for week

order_items_prod['Week'] = order_items_prod.order_purchase_timestamp.dt.isocalendar().week
order_items_prod.head()
```

```python
qty_sales_by_week = order_items_prod.groupby('Week').agg({'product_id':'count','price': 'sum'}).reset_index()
qty_sales_by_week.rename(columns={'price':'Sales', 'product_id':'Quantity',}).head()
```

```python
# Analyse sales, quantity by day. Add a new column for day

order_items_prod['Day'] = order_items_prod.order_purchase_timestamp.dt.isocalendar().day
order_items_prod.head()
```

```python
qty_sales_by_day = order_items_prod.groupby('Day').agg({'product_id':'count','price': 'sum'})
qty_sales_by_day.rename(columns={'price':'Sales', 'product_id':'Quantity',})
```

```python
# Analyse sales, quantity by payment method
#Merge orders and order_payments

pay_order=orders.merge(order_payments, on='order_id')
pay_order.head()
```

```python
qty_sales_by_paymethod = pay_order.groupby('payment_type').agg({'order_id':'count','payment_value': 'sum'}).rename(columns={'payment_value':'Sales', 'order_id':'Quantity'})
qty_sales_by_paymethod
```

```python
# Analyse sales, quantity by location

cust_by_loc=customers.merge(geo_loc, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix')
cust_by_loc
```

```python
cust_order_by_loc=cust_by_loc.merge(order_items_prod, on='customer_id')
cust_order_by_loc
```

```python
# locationwise sales and quantity
qty_sales_by_loc = cust_order_by_loc.groupby('customer_state').agg({'order_id':'count','price': 'sum'})
qty_sales_by_loc.rename(columns={'price':'Sales', 'order_id':'Quantity'})
```

#### f. Popular Products by month, seller, state, category.

```python
# Popular products by seller

prod_seller = order_items_prod.groupby(['seller_id','product_id']).agg({'product_id':'count'}).rename(columns={'product_id':'count'}).reset_index()
prod_seller = prod_seller.sort_values(by=['seller_id','count']).drop_duplicates(subset='seller_id',keep='last')
prod_seller.head()
```

```python
# Popular products by category

prod_cat = order_items_prod.groupby(['product_category_name','product_id']).agg({'product_id':'count'}).rename(columns={'product_id':'count'}).reset_index()
prod_cat = prod_cat.sort_values(by=['product_category_name','count']).drop_duplicates(subset='product_category_name',keep='last')
prod_cat.head()
```

```python
# Pouplar products by location

prod_loc = cust_order_by_loc.groupby(['customer_state','product_id']).agg({'product_id':'count'}).rename(columns={'product_id':'Count'}).reset_index()

prod_loc= prod_loc.sort_values(by=['customer_state', 'Count']).drop_duplicates(subset='customer_state', keep='last')
prod_loc
```

```python
# Popular products by Month
prod_month = order_items_prod.groupby(['month','product_id']).agg({'product_id':'count'}).rename(columns={'product_id':'count'}).reset_index()

prod_month= prod_month.sort_values(by=['month','count']).drop_duplicates(subset='month', keep='last')

prod_month
```

#### g. Popular categories by state, month

```python
# Popular category by state
cat_by_loc = cust_order_by_loc.groupby(['customer_state','product_category_name']).agg({'product_id':'count'}).rename(columns={'product_id':'Count'}).reset_index()

cat_by_loc= cat_by_loc.sort_values(by=['customer_state', 'Count']).drop_duplicates(subset='customer_state', keep='last')
cat_by_loc
```

```python
# Popular category by month
cat_by_month = cust_order_by_loc.groupby(['month','product_category_name']).agg({'product_category_name':'count'}).rename(columns={'product_category_name':'Count'}).reset_index()

cat_by_month= cat_by_month.sort_values(by=['month', 'Count']).drop_duplicates(subset='month', keep='last')
cat_by_month
```

### 3. Cross-Selling (Which products are selling together)

```python
orders_order_items = pd.merge(orders , order_items , how='left' , on='order_id' )
orders_order_items.head(2)
```

```python
cross_selling = orders_order_items[orders_order_items.duplicated(['order_purchase_timestamp','customer_id'],
                                                                 keep=False)].sort_values(by='order_purchase_timestamp')
cross_selling = cross_selling.loc[:,['customer_id','order_purchase_timestamp','product_id' ]]

cross_selling.drop_duplicates(subset ="product_id" ,inplace=True)
cross_selling = cross_selling[cross_selling.duplicated(['order_purchase_timestamp','customer_id'],
                                                       keep=False)].sort_values(by='order_purchase_timestamp').reset_index()
cross_selling.drop(columns = 'index',axis=1, inplace=True)

cross_selling
```

```python
print('Products with same customer id and order purchase timestamp are cross selling:',cross_selling.shape[0])
```
